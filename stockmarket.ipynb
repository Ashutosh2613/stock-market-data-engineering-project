```python
import pandas as pd
import requests
```

```python
#stock market data for year 2009-01 (stock = ibm)
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=IBM&interval=5min&month=2009-01&outputsize=full&apikey=BGPEQ8T5S9GS5KE9'
```

```python
print(url)
```

```python
req = requests.get(url)
```

```python
data = req.json()
```

```python
timestamp = '2009-01-20 09:50:00'
```

```python
 data['Meta Data']['2. Symbol']
```

```python
data_list = []
stock =  data['Meta Data']['2. Symbol']
for timestamp,value in data['Time Series (5min)'].items():
  timestamp = timestamp
  open = value['1. open']
  close  =  value['4. close']
  low = value['3. low']
  high = value['4. close']
  volume = value['5. volume']
  data_dict = {'stock':stock,'timestamp':timestamp,'open':open,'close':close,'low':low,'high':high,'volume':volume}
  data_list.append(data_dict)
```

```python
data_df = pd.DataFrame(data_list)
```
